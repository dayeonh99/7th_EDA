```python
df_2016.rename(columns = {'Economy_(GDP_per_Capita)' : 'Economy'}, inplace = True)
df_2016.rename(columns = {'Health_(Life_Expectancy)' : 'Health'}, inplace = True)
df_2016.rename(columns = {'Trust_(Government_Corruption)' : 'Trust'}, inplace = True)
df_2016.drop('Lower_Confidence_Interval', inplace=True, axis=1)
df_2016.drop('Upper_Confidence_Interval', inplace=True, axis=1)
df_2016.drop('Dystopia_Residual', inplace=True, axis=1)
```

```python
df_2018.rename(columns = {'Overall_rank' : 'Happiness_Rank'}, inplace = True)
df_2018.rename(columns = {'Country_or_region' : 'Country'}, inplace = True)
df_2018.rename(columns = {'Score' : 'Happiness_Score'}, inplace = True)
df_2018.rename(columns = {'GDP_per_capita' : 'Economy'}, inplace = True)
df_2018.rename(columns = {'Social_support' : 'Family'}, inplace = True)
df_2018.rename(columns = {'Healthy_life_expectancy' : 'Health'}, inplace = True)
df_2018.rename(columns = {'Freedom_to_make_life_choices' : 'Freedom'}, inplace = True)
df_2018.rename(columns = {'Perceptions_of_corruption' : 'Trust'}, inplace = True)
```

# 연도별 데이터 합치기: df_concat

```python
df_2015['Year']=2015
df_2016['Year']=2016
df_2017['Year']=2017
df_2018['Year']=2018
df_2019['Year']=2019

df_concat = pd.concat([df_2015,df_2016,df_2017, df_2018, df_2019], axis=0, sort = False)

df_concat.head()
```

#Region 결측치 채우기

```python
df_concat.isna().sum()
```

```python
df_concat = df_concat.sort_values(["Country", "Year"], ascending=True)
df_concat["Region"] = df_concat["Region"].fillna(method="ffill")
df_concat = df_concat.sort_values(["Year", "Happiness_Rank"], ascending=True)

df_concat.head()
```

```python
country_dict = dict(df_concat["Country"].value_counts())

country_check = []

for country in country_dict:
    if country_dict[country] <= 3:
        country_check.append(country)

country_check
```

```python
df_concat.loc[df_concat["Country"] == 'Trinidad & Tobago', "Region"]  = 'North America'
df_concat.loc[df_concat["Country"] == 'Northern Cyprus', "Region"] = "Western Europe"
df_concat.loc[df_concat["Country"] == 'Hong Kong S.A.R., China', "Region"] = 'Eastern Asia'
df_concat.loc[df_concat["Country"] == 'Taiwan Province of China', "Region"] = 'Eastern Asia'
df_concat.loc[df_concat["Country"] == 'North Macedonia', "Region"] = 'Western Europe'
df_concat.loc[df_concat["Country"] == 'Gambia', "Region"] = 'Sub-Saharan Africa'
```

```python
df_concat.isna().sum()
```

## df_region_2: df_mean 내에서 각 Region의 분포

```python
#df_mean에 Region 추가
df_mean.reset_index(level="Country", inplace=True)
df1 = df_mean["Country"]
df2 = df_concat[["Country", "Region"]]
df3 = pd.merge(df1, df2, left_on='Country', right_on='Country', how='left').drop_duplicates(["Country", "Region"])
df_mean = pd.merge(df_mean, df3, left_on="Country", right_on="Country", how="left")
df_mean
```

```python
df_region_2 = df_mean.groupby(["Region"], as_index=False).mean()
df_region_2 = df_region_2.sort_values("Happiness_Rank", ascending=True)
df_region_2 = df_region_2.drop("Year", axis=1)
df_region_2
```

## 각 Region 내에서 Country의 분포

```python
df_Western_Europe = df_mean[df_mean["Region"] == 'Western Europe']
df_North_America = df_mean[df_mean["Region"] == 'North America']
df_Australia_and_New_Zealand = df_mean[df_mean["Region"] == 'Australia and New Zealand']
df_Latin_America_and_Caribbean = df_mean[df_mean["Region"] == 'Latin America and Caribbean']
df_Middle_East_and_Northern_Africa = df_mean[df_mean["Region"] == 'Middle East and Northern Africa']
df_Central_and_Eastern_Europe = df_mean[df_mean["Region"] == 'Central and Eastern Europe']
df_Southeastern_Asia = df_mean[df_mean["Region"] == 'Southeastern Asia']
df_Eastern_Asia = df_mean[df_mean["Region"] == 'Eastern Asia']
df_Sub_Saharan_Africa = df_mean[df_mean["Region"] == 'Sub-Saharan Africa']
df_Southern_Asia = df_mean[df_mean["Region"] == 'Southern Asia']

df_region_list = [df_Western_Europe, df_North_America, df_Australia_and_New_Zealand, 
                  df_Latin_America_and_Caribbean, df_Middle_East_and_Northern_Africa,
                  df_Central_and_Eastern_Europe, df_Southeastern_Asia, df_Eastern_Asia,
                  df_Sub_Saharan_Africa, df_Southern_Asia]

region_name_list = ['Western Europe', 'North America', 'Australia and New Zealand',
       'Latin America and Caribbean', 'Middle East and Northern Africa',
       'Central and Eastern Europe', 'Southeastern Asia', 'Eastern Asia',
       'Sub-Saharan Africa', 'Southern Asia']
```

```python
for i, df in enumerate(df_region_list):
    df.plot.bar(x="Country", y="Happiness_Rank", title=f"{region_name_list[i]} - Happiness_Rank")
    df.plot.line(x="Country", y=["Economy", "Family", "Health", "Freedom", "Trust", "Generosity"], title=f"{region_name_list[i]} - other features")
    plt.xticks(rotation=45)
```
