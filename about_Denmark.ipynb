```python
# df_2020, df_2021에 Happiness_Rank column 추가
df_2020.sort_values("Happiness_Score", ascending=False, inplace=True)
df_2020['Happiness_Rank'] = list(df_2020.index)
df_2020['Happiness_Rank'] += 1

df_2021.sort_values("Happiness_Score", ascending=False, inplace=True)
df_2021['Happiness_Rank'] = list(df_2021.index)
df_2021['Happiness_Rank'] += 1
```

```python
# df_2017, df_2018, df_2019에 Region column 추가
dic = {}

for i in range(158):
    dic[f'{df_2015.iloc[i,0]}']=f'{df_2015.iloc[i,1]}'

for i in range(157):
    if df_2016.iloc[i,0] in dic:
        pass
    else:
        dic[f'{df_2016.iloc[i,0]}']=f'{df_2016.iloc[i,1]}'

for i in range(153):
    if df_2020.iloc[i,0] in dic:
        pass
    else:
        dic[f'{df_2020.iloc[i,0]}']=f'{df_2020.iloc[i,1]}'

for i in range(149):
    if df_2021.iloc[i,0] in dic:
        pass
    else:
        dic[f'{df_2021.iloc[i,0]}']=f'{df_2021.iloc[i,1]}'
```

# 1. Denmark

```python
df_Denmark = df_scaled[df_scaled['Country'] == "Denmark"]
df_Denmark
```

```python
# 변수 간 상관관계
# Happiness Rank, Happiness Score 이상해.. => 표본이 너무 작아서인듯? 밑에 연도에 따른 변화에서 Rank와 Score의 경향성 일치하지 X
plt.rcParams['figure.figsize']=6,5

df = df_Denmark.drop("Year", axis=1)
_ = sns.heatmap(df.corr().abs(), annot=True, fmt=".2f", cmap="Purples", linewidths=1)
```

- 덴마크의 Happiness Score은 Generosity와 강한 상관관계를 가지며, Health, Freedom, Trust와도 상관관계가 있다.
<br> -> 전세계적으로 Happiness Score가 Economy, Social_support, Health와 강한 상관관계를 가지는 것과는 다른 양상이다.
- Economy, Social_support, Health의 상관관계가 전세계적인 경향과 일치하나, 그 정도는 평균 이하이다.
- Social_support와 Health, Social_support와 Trust, Freedom과 Trust, Freedom과 Generosity가 0.4 이상의 상관관계를 가진다
<br> -> Happiness Score와 비교적 높은 상관관계를 가지는 변수들은 대체적으로 그 변수들끼리도 높은 상관관계를 가진다.


```python
# 연도에 따른 변화
f, axes = plt.subplots(1, 3, figsize=(15,4))

plt.subplot(1, 3, 1)
plt.plot(df_Denmark.Year, df_Denmark.Happiness_Rank, marker='o', color='blue')
plt.title('Happiness Rank of Denmark', fontsize=14)

plt.subplot(1, 3, 2)
plt.plot(df_Denmark.Year, df_Denmark.Happiness_Score, marker='o', color='blue')
plt.title('Happiness Score of Denmark', fontsize=14)

plt.subplot(1, 3, 3)
plt.plot(df_Denmark.Year, df_Denmark.Economy, marker='o', color='green')
plt.plot(df_Denmark.Year, df_Denmark.Social_support, marker='o', color='blue')
plt.plot(df_Denmark.Year, df_Denmark.Health, marker='o', color='yellow')
plt.plot(df_Denmark.Year, df_Denmark.Freedom, marker='o', color='red')
plt.plot(df_Denmark.Year, df_Denmark.Trust, marker='o', color='grey')
plt.plot(df_Denmark.Year, df_Denmark.Generosity, marker='o', color='purple')
plt.title('Other Features of Denmark', fontsize=14)
plt.legend(["Economy", "Social_support", "Health", "Freedom", "Trust", "Generosity"], fontsize=12, bbox_to_anchor = (1,1))

plt.show()
```

- 덴마크의 Happiness Rank는 1~3위 사이로 비교적 일정하므로, Happiness Score을 주된 지표로 삼는 것이 바람직해 보인다.
- 위 상관계수표를 통한 예측과 다르게, Happiness_Score와 Generosity, Health, Freedom, Trust의 경향성은 비슷하지 않다. (왜...?)
